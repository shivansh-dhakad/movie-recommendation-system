- converting string of list to list 

```python
import ast
```

- getting genres in a list

```python
def get_name(x):
    l=[]
    for i in ast.literal_eval(x):
        l.append(i['name'])
    return l
```

```python
df['genres']=df['genres'].apply(get_name)
```

```python
df['genres'][0]
```

- converting keyword into a list of strings

```python
df['keywords'][1]
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i)
    return l
```

```python
df['keywords']=df['keywords'].apply(convert)
```

- getting top 4 cast

```python
df['cast'][0]
```

```python
def cast_convert(x):
    l=[]
    ct=0
    for i in ast.literal_eval(x):
        if ct!=4:
            l.append(i['name'])
            ct+=1
        else:
            break
    return l
```

```python
df['cast']=df['cast'].apply(cast_convert)
```

- gettiing director from crew

```python
df['crew'][0]
```

```python
def fetch_director(x):
    l=[]
    for i in ast.literal_eval(x):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
df['crew']=df['crew'].apply(fetch_director)
```

- removing spaces from genres,keyword,cast and crew

```python
df['genres']=df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['keywords']=df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast']=df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew']=df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# converting overview into list of strings
df['overview']=df['overview'].apply(lambda x:x.split())
```

- creating a new column named tags

```python
df['tags']=df['overview']+df['genres']+df['keywords']+df['cast']+df['crew']
```

- creating a new dataset

```python
new_df=df[['movie_id','title','tags']]
```

```python
new_df.sample()
```

```python
new_df=new_df.reset_index()
```

- converting tags into a single para in lower case

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

- cleaning tags with cleantext library

```python
import cleantext
```

```python
# clean the text
def clean_text(data):
    cleaned_text=cleantext.clean(
        data,
        extra_spaces=True, #remove extra spaces
        stemming=False, #apply stremming(ex:"running" to "run")
        stopwords=True, #remove stopwords(is,a,the)
        lowercase=True, #convert text to lowercase
        numbers=True, #remove all digits
        punct=True #remove all punctuation
    )
    return cleaned_text
```

```python
new_df['tags']=new_df['tags'].apply(clean_text)
```

```python
new_df.iloc[0]['tags']
```

- converting words to base word(ex:running to run)

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmetizer=WordNetLemmatizer()
```

```python
def stem(text):
    l=[]
    for i in text.split():
        l.append(lemmetizer.lemmatize(i,pos='v'))
    return " ".join(l)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.iloc[0]['tags']
```

-  converting words to array of vector

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words='english',max_features=5000)
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

- getting cosine similarity between vectors rowwise

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity.shape
```

- sample recommendor system

```python
user_input=input("search movie:")
try:
    movie_index=new_df[new_df['title'].str.lower()==user_input.lower()].index[0]
    distances=similarity[movie_index]
    # sorting distances according to distance
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:10]
    # print(movie_list)
    print(f"------------movies similar to {user_input}-----------------")
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
except Exception as e:
    print('Error:',e)
```

```python
import pickle
with open('similarities.pkl','wb') as file:
    pickle.dump(similarity,file)
new_df.to_csv('new_df.csv')
```
